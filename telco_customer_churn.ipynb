```python
# Adapt 'TotalCharges' data type
df_adapted = df.copy()
df_adapted['TotalCharges'] = pd.to_numeric(df_adapted['TotalCharges'], errors='coerce')

# Check for missing values
count_missing_values = df_adapted['TotalCharges'].isna().sum().item()
print(f'Missing TotalCharges: {count_missing_values}')
```
